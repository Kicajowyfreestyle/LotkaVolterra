```python
# This program is about simulation of Predator-Prey problem in discrete time
```

```python
import matplotlib.pyplot as plt
```

```python
### proposal of model with grass
alfa = 2
beta = 0.1
gamma = 0.005
delta = 0.1
epsilon = 0.005
dzeta = 0.2
eta = 0.003
G = []
Z = []
W = []
G.append(100)
Z.append(100)
#W.append(100)
W.append(0)
time = 10000
```

```python
for i in range(time):
    G.append(G[i]*alfa/(1+eta*G[i])-beta*Z[i])
    Z.append((gamma*G[i])*Z[i]-delta*W[i])
    #W.append(W[i]*(epsilon*Z[i]-dzeta))
    W.append(0)
    if Z[i+1]<=0:# or W[i+1]<=0:
        break
```

```python
ranMax = 1000
ranMin = 0
grass, = plt.plot(G[ranMin:ranMax], label='Grass')
prey, = plt.plot(Z[ranMin:ranMax], label='Rabbits')
#predator, = plt.plot(W[ranMin:ranMax], label='Predator')
plt.legend(handles=[grass, prey])
plt.show()
```
